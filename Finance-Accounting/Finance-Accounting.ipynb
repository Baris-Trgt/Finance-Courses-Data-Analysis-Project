## Data Cleaning

```python
#We delete the columns that are not useful in the dataset.
dataFrame.drop(["avg_rating_recent","discount_price__amount","price_detail__currency","price_detail__amount","discount_price__price_string"],
               axis=1, inplace=True , errors="ignore")
```

```python
#Missing Values Heatmap.  (discount_price__currency //// price_detail__price_string)
sns.heatmap(dataFrame.isnull())
plt.title("Missing Values Heatmap")
plt.show()
```

```python
plt.pie(dataFrame["is_paid"].value_counts(), labels=['Paid Courses (True)', 'Free Courses (False)'], autopct="%1.1f%%")
plt.figure(figsize=(10,10))
plt.show()
```

```python
# Replacing missing values with the average instead of removing them.
# The mean() function cannot be applied because the data type is "object".
dataFrame = dataFrame.dropna(subset=['discount_price__currency', 'price_detail__price_string'])
```

## EDA (Exploratory Data Analysis)

```python
plt.hist(dataFrame["num_subscribers"])
plt.xlabel("Courses")
plt.ylabel("Number of People Subscribed to the Course")
plt.show()
```

```python
# We defined a function that selects the most popular courses based on the number of subscribers and displays their titles (title) and average ratings (avg_rating) to the user.
def top_Course(dataFrame):
    popular = dataFrame.sort_values(by="num_subscribers",ascending=False).head(20)
    return popular[['title','avg_rating']]   
```

```python
plt.figure(figsize=(10,10))
plt.barh(top_Course(dataFrame)['title'], top_Course(dataFrame)['avg_rating'])
plt.xlabel("Average Rating")
plt.ylabel("Title")
plt.title("Top 20 Most Subscribed Courses and Their Average Ratings")
plt.show()
```

```python
# We created a function that displays the top 10 courses with the most subscribers and provides the course names and subscriber numbers to the user.
def num_subscribers_top_ten(dataFrame):
    num_subscribers = dataFrame.sort_values(by="num_subscribers",ascending=False).head(10)
    return num_subscribers[['title','num_subscribers']]   
```

```python
plt.figure(figsize=(10,10))
plt.bar(num_subscribers_top_ten(dataFrame)['title'],num_subscribers_top_ten(dataFrame)['num_subscribers'],color='skyblue')
plt.xlabel("Title")
plt.ylabel("Num Subscribers")
plt.xticks(rotation=90, ha='right')
plt.title("Top 10 Most Subscribed Courses and Their Num Subscribers")
plt.show()
```

```python
#We made a function that shows the number of courses published by year.
dataFrame['published_time'] = pd.to_datetime(dataFrame['published_time'])
dataFrame['year'] = dataFrame['published_time'].dt.year
course_counts = dataFrame.groupby('year').size()
```

```python
plt.figure(figsize=(15,7))
plt.plot(course_counts.index, course_counts.values, marker='o', linestyle='-')
plt.title('Number of Courses Published by Year')
plt.xlabel('Year')
plt.ylabel('Published Time')
plt.show()
```
